```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("data/Mental_Health_and_Social_Media_Balance_Dataset.csv")
df
```

```python
df.isna().sum()
```

```python
print(df['Gender'].value_counts())
df_ohe = pd.get_dummies(df, columns=['Gender', 'Social_Media_Platform'], dtype=int)
df_ohe
```

```python
df_modified = df_ohe.drop(columns=['User_ID'])
df_modified
```

```python
df_modified.columns
df_modified.to_csv("data/df_modified.csv",index=False)
```

```python
correlation_matrix = df_modified.corr()
correlation_matrix.sort_values(by=["Stress_Level(1-10)"],ascending=False,inplace=True)
correlation_matrix["Stress_Level(1-10)"]
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(df_modified.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
